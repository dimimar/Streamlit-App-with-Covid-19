```python
#Please run the following code to and IDE like Spyder and not in jupyter
#import libraries
import pandas as pd
import numpy as np
import altair as alt
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px  # interactive charts
import streamlit as st
#read the file that is taken from kaggle dataset COVID-19 Coronavirus.csv
file=pd.read_csv("C:/Users/gluma/Downloads/archive/COVID-19 Coronavirus.csv")
#create another dataframe that has Country as index
file1=file.set_index("Country")
#Set the main title of the page, the page name and use an icon of your preference
st.set_page_config(
    page_title="Data Science Dashboard About Covid",
    page_icon="üê¶",
    layout="wide",
)
st.title("Data Science Dashboard About Covid")
#Create a dropdown filter by Continent
Country_filter = st.selectbox("Select the Continent", pd.unique(file1["Continent"]))
file1=file1[file1["Continent"] == Country_filter]
#Use groupby method to create 2 grouped dataframes
grouped1=file1.groupby(by=["Country"]).sum([["Tot Cases//1M pop"]]).sort_values(by="Tot\xa0Cases//1M pop")
grouped2=file1.groupby(by=["Country"]).sum([["Tot Deaths/1M pop"]]).sort_values(by="Tot\xa0Deaths/1M pop")
#Use a subheader
st.subheader("Country Breakdown üìà")
#place our barcharts side by side and create the charts based on grouped dataframes created before
col1, col2 = st.columns(2)
with col1:
    chart1=px.bar(grouped1,x="Tot\xa0Cases//1M pop",y=grouped1.index,orientation="h",color_discrete_sequence=["#dc143c"]*len(grouped1),template="plotly_white",title="Total Cases/1 M Pop by Country")
    st.plotly_chart(chart1)
with col2:
    chart2=px.bar(grouped2,x="Tot\xa0Deaths/1M pop",y=grouped2.index,orientation="h",color_discrete_sequence=["#dc143c"]*len(grouped2),template="plotly_white",title="Total Deaths/1 M Pop by Country")
    st.plotly_chart(chart2)
st.dataframe(file1[["Total Cases","Total Deaths","Death percentage"]].sort_values(by="Total Deaths",ascending=False))
#Create bubblechart to plot Total Cases and Total Deaths per 1M Pop
bubblechart=alt.Chart(file).mark_circle(size=100).encode(
    x='Tot\xa0Deaths/1M pop',
    y='Tot\xa0Cases//1M pop',
    tooltip=['Country','Tot\xa0Cases//1M pop', 'Tot\xa0Deaths/1M pop'])
st.altair_chart(bubblechart,use_container_width=True)
```
